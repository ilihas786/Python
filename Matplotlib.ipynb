# Matplotlib Chart Examples

This notebook demonstrates how to use matplotlib to plot different types of charts using a sample dataset with two columns: name and marks. Each function's syntax is explained.

```python
# Import matplotlib and create a sample dataset
import matplotlib.pyplot as plt
import numpy as np

# Sample dataset: two columns (name and marks)
names = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
marks = [85, 92, 78, 88, 95]
```

## Line Chart

A line chart displays data as a series of points connected by straight lines. Use `plt.plot()` to create a line chart.

**Syntax:**

- `plt.plot(x, y)`: Plots y versus x as lines.
- `plt.title('Title')`: Sets the chart title.
- `plt.xlabel('X Label')`, `plt.ylabel('Y Label')`: Set axis labels.
- `plt.show()`: Displays the plot.

```python
# Line chart example
plt.plot(names, marks, marker='o')
plt.title('Line Chart of Marks')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.show()
```

## Bar Chart

A bar chart displays data with rectangular bars. Use `plt.plot()` with the `'bar'` style to create a bar chart.

**Syntax:**

- `plt.plot(x, y, 'o-')`: Plots y values as bars for each x value (for true bar chart, use `plt.bar()`, but here we use plot for demonstration).

- `plt.title('Title')`: Sets the chart title.
- `plt.xlabel('X Label')`, `plt.ylabel('Y Label')`: Set axis labels.
- `plt.show()`: Displays the plot.

```python
# Bar chart example using plot function
import pandas as pd
data={'names' : ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],'marks' : [85, 92, 78, 88, 95]}
df=pd.DataFrame(data,columns=['names', 'marks'])
df.plot(x='names', y='marks', kind='bar', color='orange') 
```

## Horizontal Bar Chart

A horizontal bar chart displays data with bars running horizontally. Use `df.plot(kind='barh')` to create a horizontal bar chart.

**Syntax:**

- `df.plot(x='names', y='marks', kind='barh', color='skyblue')`: Plots marks as horizontal bars for each name.
- `plt.title('Title')`: Sets the chart title.
- `plt.xlabel('X Label')`, `plt.ylabel('Y Label')`: Set axis labels.
- `plt.show()`: Displays the plot.

```python
# Horizontal bar chart example using pandas and matplotlib
df.plot(x='names', y='marks', kind='barh', color='skyblue')
plt.title('Horizontal Bar Chart of Marks')
plt.xlabel('Marks')
plt.ylabel('Name')
plt.show()
```

## Scatter Plot

A scatter plot displays individual data points. Use `df.plot(kind='scatter')` to create a scatter plot.

**Syntax:**

- `df.plot(x='names', y='marks', kind='scatter', color='red')`: Plots marks as points for each name.
- `plt.title('Title')`: Sets the chart title.
- `plt.xlabel('X Label')`, `plt.ylabel('Y Label')`: Set axis labels.
- `plt.show()`: Displays the plot.

```python
# Scatter plot example using pandas and matplotlib
df.plot(x='names', y='marks', kind='scatter', color='red')
plt.title('Scatter Plot of Marks')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.show()
```

## Area Chart

An area chart displays quantitative data graphically. Use `plt.plot()` and `plt.fill_between()` to create an area chart.

**Syntax:**

- `plt.plot(x, y)`: Plots the line.
- `plt.fill_between(x, y)`: Fills the area under the line.
- `plt.title('Title')`: Sets the chart title.
- `plt.xlabel('X Label')`, `plt.ylabel('Y Label')`: Set axis labels.
- `plt.show()`: Displays the plot.

```python
# Area chart example
plt.plot(names, marks, color='green')
plt.fill_between(names, marks, color='lightgreen', alpha=0.5)
plt.title('Area Chart of Marks')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.show()
```

## Pie Chart

A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportions. Each slice represents a category's contribution to the total. Pie charts are useful for showing percentage or proportional data.

**Syntax:**
- `plt.pie(values, labels=labels, autopct='%1.1f%%')`: Plots a pie chart with given values and labels. `autopct` displays the percentage on each slice.
- `plt.title('Title')`: Sets the chart title.
- `plt.show()`: Displays the plot.

```python
data = {'Language': ['Python', 'Java', 'C++', 'JavaScript', 'Ruby'],
        'Popularity': [85, 75, 65, 80, 70]}
df = pd.DataFrame(data,columns=['Language', 'Popularity'])
df.plot.pie(y='Popularity', labels=df['Language'])
```
