## Join Operations in pandas
Pandas provides several types of join operations to combine DataFrames based on a common column (key). The most common join types are:

- **Inner Join**: Returns only the rows with keys present in both DataFrames.
- **Left Join**: Returns all rows from the left DataFrame and matching rows from the right DataFrame. Non-matching rows from the right will have NaN.
- **Right Join**: Returns all rows from the right DataFrame and matching rows from the left DataFrame. Non-matching rows from the left will have NaN.
- **Outer Join**: Returns all rows from both DataFrames. Non-matching rows will have NaN in the columns where data is missing.

Let's see an example with two small DataFrames:

````python
# Example DataFrames for join operations
df1 = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']
})

df2 = pd.DataFrame({
    'id': [3, 4, 5, 6],
    'score': [85, 90, 95, 80]
})

# Inner Join
inner = pd.merge(df1, df2, on='id', how='inner')
print('Inner Join:')
print(inner)

# Left Join
left = pd.merge(df1, df2, on='id', how='left')
print('\nLeft Join:')
print(left)

# Right Join
right = pd.merge(df1, df2, on='id', how='right')
print('\nRight Join:')
print(right)

# Outer Join
outer = pd.merge(df1, df2, on='id', how='outer')
print('\nOuter Join:')
print(outer)
````

## Concat Operation in pandas

The `concat` function in pandas is used to combine DataFrames either vertically (stacking rows) or horizontally (adding columns).

- **Vertical Concatenation**: Use `pd.concat([df1, df2])` to stack DataFrames on top of each other (default axis=0).

- **Horizontal Concatenation**: Use `pd.concat([df1, df2], axis=1)` to add columns side by side.

````python
# Example of vertical concatenation (stacking rows)
df3 = pd.DataFrame({
    'id': [7, 8],
    'name': ['Eve', 'Frank']
})

vertical_concat = pd.concat([df1, df3], ignore_index=True)
print('Vertical Concatenation:')
print(vertical_concat)

# Example of horizontal concatenation (adding columns)
df4 = pd.DataFrame({
    'age': [25, 30, 35, 40]
})

horizontal_concat = pd.concat([df1, df4], axis=1)
print('\nHorizontal Concatenation:') 
print(horizontal_concat)
````

## Difference between loc and iloc in pandas

- **loc** is label-based: it is used to access rows and columns by their labels (names).

- **iloc** is integer position-based: it is used to access rows and columns by their integer positions (like Python list indexing).

### Examples:

Suppose you have a DataFrame with custom row labels:

```python
df = pd.DataFrame({'A': [10, 20, 30]}, index=['x', 'y', 'z'])

```

- `df.loc['x']` returns the row with label 'x'.

- `df.iloc[0]` returns the first row (regardless of its label).

- `df.loc['x':'y']` selects rows from 'x' to 'y' (inclusive).

- `df.iloc[0:2]` selects the first two rows (like list slicing, exclusive of the stop index).
