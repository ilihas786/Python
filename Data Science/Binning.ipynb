```python
import pandas as pd
```

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df
```

```python
# Bining For Age Columns
df.Age.max()
```

```python
df.Age.min()
```

```python
labels=['0-20','21-40','41-60','above 61'] # create lables first
bins =[0,21,41,61,100] # Always take first value of lables
df['bins']=pd.cut(df.Age,bins,labels=labels)
```

```python
df[['Age','bins']]
```

```python
# Plotting The Graph
import matplotlib.pyplot as plt
plt.bar(labels,df['bins'].value_counts())
plt.title('Age Groups VS Count')

plt.xlabel('Age Groups')
plt.ylabel('Count')

plt.show()
```
